```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("creditcard.csv")

# Preview data
print(df.head())

# Check data info
df.info()

# Descriptive stats
print(df[['Time', 'Amount', 'Class']].describe())

# Class distribution
class_counts = df['Class'].value_counts()
class_percent = (class_counts / len(df)) * 100
print(class_counts)
print(class_percent)
```

```python
# Plot 1: Class Distribution (Fraud vs Genuine)
sns.countplot(x='Class', data=df, palette=['skyblue', 'salmon'])
plt.title(" Class Distribution of Transactions")
plt.xlabel("Transaction Class (0 = Genuine, 1 = Fraud)")
plt.ylabel("Number of Transactions")
plt.show()

# Plot 2: Amount Distribution by Class
sns.histplot(data=df, x='Amount', hue='Class', bins=100, palette=['blue', 'red'], alpha=0.6)
plt.title(" Transaction Amount Distribution by Class")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.xlim(0, 2500)  # Optional: zoom to focus on most transactions
plt.show()

# Plot 3: Time Distribution by Class
sns.histplot(data=df, x='Time', hue='Class', bins=100, palette=['green', 'orange'], alpha=0.6)
plt.title(" Transaction Time Distribution by Class")
plt.xlabel("Time (seconds since first transaction)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Load data
df = pd.read_csv("creditcard.csv")

# Class distribution bar plot
sns.countplot(data=df, x='Class', palette=['skyblue', 'salmon'])
plt.title("Figure 1: Class Distribution of Transactions")
plt.xlabel("Transaction Class (0 = Genuine, 1 = Fraud)")
plt.ylabel("Number of Transactions")
plt.xticks([0, 1], ['Genuine', 'Fraud'])
plt.show()

# Histogram of transaction amount by class
sns.histplot(data=df, x='Amount', hue='Class', bins=100, palette=['blue', 'red'], alpha=0.6)
plt.title("Figure 2: Transaction Amount Distribution by Class")
plt.xlabel("Transaction Amount (€)")
plt.ylabel("Frequency")
plt.xlim(0, 2500)  # Focus on bulk of transactions
plt.legend(title='Class', labels=['Genuine', 'Fraud'])
plt.show()


# Histogram of transaction time by class
sns.histplot(data=df, x='Time', hue='Class', bins=100, palette=['green', 'orange'], alpha=0.6)
plt.title("Figure 3: Transaction Time Distribution by Class")
plt.xlabel("Time (seconds since first transaction)")
plt.ylabel("Frequency")
plt.legend(title='Class', labels=['Genuine', 'Fraud'])
plt.show()

# Boxplot of transaction amounts
sns.boxplot(data=df, x='Class', y='Amount', palette=['skyblue', 'salmon'])
plt.title("Figure 4: Boxplot of Transaction Amounts by Class")
plt.xlabel("Transaction Class")
plt.ylabel("Transaction Amount (€)")
plt.xticks([0, 1], ['Genuine', 'Fraud'])
plt.ylim(0, 3000)  # Limit Y for better view; you can adjust/remove this
plt.show()
```

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Load the dataset
df = pd.read_csv("creditcard.csv")

# Plot: Class Distribution of Transactions
sns.countplot(data=df, x='Class', palette=['dodgerblue', 'tomato'])
plt.title("Class Distribution of Transactions")
plt.xlabel("Transaction Class (0 = Genuine, 1 = Fraud)")
plt.ylabel("Number of Transactions")
plt.xticks([0, 1], ['Genuine', 'Fraud'])
plt.show()

#  Plot: Transaction Amount Distribution by Class
sns.histplot(data=df, x='Amount', hue='Class', bins=100, palette=['seagreen', 'crimson'], alpha=0.7)
plt.title("Transaction Amount Distribution by Class")
plt.xlabel("Transaction Amount (€)")
plt.ylabel("Frequency")
plt.xlim(0, 2500)
plt.legend(title='Class', labels=['Genuine', 'Fraud'])
plt.show()

# Plot: Transaction Time Distribution by Class
sns.histplot(data=df, x='Time', hue='Class', bins=100, palette=['goldenrod', 'steelblue'], alpha=0.7)
plt.title("Transaction Time Distribution by Class")
plt.xlabel("Time (seconds since first transaction)")
plt.ylabel("Frequency")
plt.legend(title='Class', labels=['Genuine', 'Fraud'])
plt.show()

# Plot: Boxplot of Transaction Amounts by Class
sns.boxplot(data=df, x='Class', y='Amount', palette=['lightgreen', 'coral'])
plt.title("Boxplot of Transaction Amounts by Class")
plt.xlabel("Transaction Class")
plt.ylabel("Transaction Amount (€)")
plt.xticks([0, 1], ['Genuine', 'Fraud'])
plt.ylim(0, 3000)
plt.show()
```
